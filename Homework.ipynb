#### 2. Построить график производной функции (2п+3п)


```python
# ищем производную

def diff(x, f):
    len_arr = len(f)
    if len(x) != len_arr:
        return None
    array_diff = np.zeros(len_arr)
    for i in range(len_arr-1):
        array_diff[i] = (f[i+1] - f[i])/(x[i+1]-x[i])
    array_diff[len_arr-1] = array_diff[len_arr - 2]
    return array_diff
        
Dif = diff(f,x)

plt.style.use('seaborn')    
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title('Производная функции', fontsize = 18)
ax.set_xlabel("x", fontsize = 12)
ax.set_ylabel("Производная", fontsize = 12)
ax.plot(x,Dif, color="yellow")

plt.show()
```

##### 5.Ищем интеграл на заданном отрезке.

```python
# будем использовать метод прямоугольников
def work(f, a, b, n):
    h = (b-a)/float(n)
    total = sum([f((a + (k*h))) for k in range(0, n)])
    result = h * total
    return result


def f(x):
    return pow(np.sin(x),4)*pow(np.cos(x),4)


n = 2
a1 = work(f, 0, 2*math.pi, n)
n *= 2
a2 = work(f, 0, 2*math.pi, n)

while abs(a1 - a2) > 0.001:
    n *= 2
    a1 = work(f, 0, 2*math.pi, n)
    n *= 2
    a2 = work(f, 0, 2*math.pi, n)

print("\nOtvet:", a2)
```
